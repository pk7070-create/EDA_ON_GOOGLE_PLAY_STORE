```python
#removing the index[10472]
df_copy=df.copy()
df_copy.head()
df_copy=df_copy.drop(df_copy.index[10472])
df_copy[~df['Reviews'].str.isnumeric()]
```

```python
#converting data_type into int.
df_copy['Reviews']=df_copy['Reviews'].astype(int)
df_copy.info()
```

```python
df['Size'].isnull().sum()
```

```python
df['Size'].unique()
```

```python



df_copy['Size']=df_copy['Size'].str.replace('M','000')
df_copy['Size']=df_copy['Size'].str.replace('k','')
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
df_copy['Size'] = pd.to_numeric(df_copy['Size'], errors='coerce')

df_copy['Size']
```

```python
df_copy.info()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python

char_to_remove=['+',',','$','Everyone','Free']
col_engaged=['Installs','Price']
for item in char_to_remove:
    for colmns in col_engaged:
        df_copy[colmns]=df_copy[colmns].str.replace(item,'')

```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs'].unique()
```

```python





df_copy['Price'].unique()
df_copy['Installs'].unique()
df_copy['Installs']=df_copy['Installs'].astype('float')
df_copy['Price']=df_copy['Price'].astype('float')


df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'],errors='coerce')
df_copy['Last Updated']









```

```python
df_copy['Last Updated']

df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Last Updated']

df_copy['day']=df_copy['Last Updated'].dt.day
df_copy['month']=df_copy['Last Updated'].dt.month
df_copy['year']=df_copy['Last Updated'].dt.year





df_copy
```

```python
#saving a cleaned data_set in the data folder.
df_copy.to_csv("data/google_cleaned.csv")
```

```python
df_copy.head()
df_copy.duplicated('App')
df_copy=df_copy.drop_duplicates(subset='App',keep='first')
df_copy.shape


#want to know the numerical and column feature in the dataset.
Numerical_Feature=[
    feature for feature in df_copy.columns if df_copy[feature].dtype!='O'
]

categorical_features=[
    feature for feature in df_copy.columns if df_copy[feature].dtype=='O'
]

print("THERE ARE {} AND THE NAMES ARE {}".format(len(Numerical_Feature),Numerical_Feature))

print("\nTHERE ARE {} AND THE NAMES ARE {}".format(len(categorical_features),categorical_features))
```

```python
#calucating the proportion of the categorical features.
#proportion = kitna percent 

for colns in categorical_features:
    print(df_copy[colns].value_counts(normalize=True)*100)
```
